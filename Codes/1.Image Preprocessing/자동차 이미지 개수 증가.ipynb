```python
from keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.utils import load_img,array_to_img,img_to_array

import numpy as np
import matplotlib.pyplot as plt
import os
import PIL
import tensorflow as tf
```

```python
data_datagen = ImageDataGenerator(rescale=1./255)
 
data_datagen = ImageDataGenerator(rescale=1./255,
                                   rotation_range=20, 
                                   shear_range=0.5,  
                                   # width_shift_range=0.1,
                                   # height_shift_range=0.1,
                                   horizontal_flip=True,
                                   vertical_flip=True,
                                   fill_mode='nearest') 
 
# model = ['Rexton','Tivoli','Torres','KorandoEmotion','BOLT_EUV','BOLT_EV','G70','G80','G90','GV60','GV70','GV80','Malibu','Spark','Tahoe','Traverse','Tahoe','Trax'] 
filename_in_dir = [] 
model1 = ['GV60','GV80','KorandoEmotion','Malibu','Torres','Traverse','Trax']
model2 = ['GV80']
for carname in model2:
    for root, dirs, files in os.walk(f'C:/Users/Gunhee/test/LASTPRO/Tenmodelcar/{carname}/'):
        for  fname in files:
            full_fname = os.path.join(root, fname)
            filename_in_dir.append(full_fname)
    for file_image in filename_in_dir:
        print (file_image)
        img = load_img(file_image) 
        x = img_to_array(img)
        x = x.reshape((1,) + x.shape)

        i = 0

        for batch in data_datagen.flow(x,save_to_dir=f'C:/Users/Gunhee/test/LASTPRO/CARTENADD2900/{carname}/', save_prefix=f'{carname}', save_format='jpg'):
            i += 1
            if i > 6:
                break
```
